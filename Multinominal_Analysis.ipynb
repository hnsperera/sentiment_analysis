```python
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
from time import sleep
from test import getData

import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.classify import NaiveBayesClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cross_validation import train_test_split
from sklearn import naive_bayes
import numpy as np
from sklearn import preprocessing
from time import sleep
```

```python
sentiment_data = pd.read_csv('/home/sampath-pc/Documents/filePositiveNegative/trainingdata.txt', sep='\t', names=['Class','Text'])
```

```python
sentiment_data
```

```python
stop_word = set(stopwords.words('english'))
vectorizer =TfidfVectorizer(use_idf=True,lowercase=True,strip_accents='ascii',stop_words=stop_word)
```

```python
y=sentiment_data.Class
```

```python
print(y)
```

```python
x=vectorizer.fit_transform(sentiment_data.Text)
```

```python
print(x)
```

```python
print (y.shape)
print (x.shape)
```

```python
#Test train spllit
X_train, X_test, Y_train, Y_test = train_test_split(x,y,random_state=42)
scaler = preprocessing.StandardScaler(with_mean=False).fit(X_train)
X_train_transformed = scaler.transform(X_train)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
#train
clf= naive_bayes.MultinomialNB()
clf.fit(X_train,Y_train)
Y_prediction=clf.predict(X_test)
```

```python
Y_prediction
```

```python
#model accuracy
print("Test accuracy is : ")


X_test_transformed = scaler.transform(X_test)
clf.score(X_test_transformed, Y_test) 
```
